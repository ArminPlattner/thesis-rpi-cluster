```python
import os
import matplotlib.pyplot as plt
import numpy as np

# --- UPDATED DATA POINTS ---
volumes = ['100 Logs', '500 Logs', '1000 Logs']

# Total processing duration (s) from your latest run:
durations = [20.43, 101.99, 207.38] 

# Consumer throughput (logs/s) from your latest run:
throughputs = [4.90, 4.90, 4.84]

# Mean inference latency (ms) from your latest run:
latencies = [198.72, 199.19, 202.31] 

# Note: Your latest Pi1 ingestion throughputs (~129.x logs/s)
ingest_throughputs = [129.23, 129.77, 129.97]


plt.rcParams['font.family'] = 'serif'

# --- CHART 1: DURATION & THROUGHPUT ---
fig, ax1 = plt.subplots(figsize=(8, 4.5), dpi=300)
color = '#1f77b4'
ax1.set_xlabel('Test Volume (Number of Logs)', fontweight='bold', labelpad=10)
ax1.set_ylabel('Total Processing Duration (seconds)', color=color, fontweight='bold')
line1 = ax1.plot(volumes, durations, marker='o', color=color, linewidth=2, label='Total Duration (s)')
ax1.tick_params(axis='y', labelcolor=color)
ax1.grid(True, linestyle='--', alpha=0.5)

ax2 = ax1.twinx()
color = '#2ca02c'
ax2.set_ylabel('Consumer Throughput (logs/sec)', color=color, fontweight='bold')
line2 = ax2.plot(volumes, throughputs, marker='s', color=color, linewidth=2, linestyle='--', label='Throughput (logs/s)')
ax2.tick_params(axis='y', labelcolor=color)
# Adjusted limits if needed, based on your new throughput range
ax2.set_ylim(4.8, 5.0) 

plt.title('End-to-End Processing Duration & System Throughput (DRAIN+INT8)', fontsize=12, fontweight='bold', pad=15)
lines = line1 + line2
labels = [l.get_label() for l in lines]
ax1.legend(lines, labels, loc='upper left')
plt.tight_layout()
plt.savefig('scale_metrics_duration_throughput_v6_drain.png', dpi=300)
plt.close()

# --- CHART 2: INFERENCE LATENCY STABILITY ---
fig, ax = plt.subplots(figsize=(7, 4), dpi=300)
ax.bar(volumes, latencies, color='#d62728', alpha=0.8, width=0.4, edgecolor='black', zorder=3)
ax.set_ylabel('Mean Inference Latency (ms)', fontweight='bold')
ax.set_xlabel('Test Volume (Number of Logs)', fontweight='bold', labelpad=10)
ax.set_ylim(0, 250)
ax.grid(True, linestyle='--', alpha=0.5, axis='y', zorder=0)
for i, v in enumerate(latencies):
    ax.text(i, v + 5, f"{v:.2f} ms", ha='center', fontweight='bold', color='#333333')
plt.title('DRAIN+INT8 Inference Latency Stability', fontsize=12, fontweight='bold', pad=15)
plt.tight_layout()
plt.savefig('scale_metrics_inference_latency_v6_drain.png', dpi=300)
plt.close()
```
