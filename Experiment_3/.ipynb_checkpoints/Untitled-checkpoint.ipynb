```python
import matplotlib.pyplot as plt

def parse_nmon(filename):
    cpu_data = []
    mem_data = []

    section = None

    with open(filename, "r") as f:
        for line in f:
            line = line.strip()
            if not line or line.startswith("#"):
                continue

            parts = line.split(",")
            if len(parts) < 2:
                continue

            kind = parts[0]

            # Section headers
            if kind == "CPU":
                section = "CPU"
                continue
            elif kind == "MEM":
                section = "MEM"
                continue
            else:
                if section == "CPU" and kind == "CPU":
                    # CPU,timestamp,sys,user,idle,wait,...
                    try:
                        ts = float(parts[1])
                        sys_c = float(parts[2])
                        user_c = float(parts[3])
                        idle_c = float(parts[4])
                        cpu_data.append((ts, sys_c, user_c, idle_c))
                    except (ValueError, IndexError):
                        pass
                elif section == "MEM" and kind == "MEM":
                    # MEM,timestamp,total,used,free,buffers,cached,...
                    try:
                        ts = float(parts[1])
                        total = float(parts[2])
                        used = float(parts[3])
                        free = float(parts[4])
                        mem_data.append((ts, total, used, free))
                    except (ValueError, IndexError):
                        pass

    return cpu_data, mem_data

def plot_cpu(cpu_data):
    if not cpu_data:
        print("No CPU data found.")
        return

    times = [r[0] for r in cpu_data]
    sys_cpu = [r[1] for r in cpu_data]
    user_cpu = [r[2] for r in cpu_data]
    idle_cpu = [r[3] for r in cpu_data]

    plt.figure(figsize=(10, 6))
    plt.plot(times, sys_cpu, label="sys")
    plt.plot(times, user_cpu, label="user")
    plt.plot(times, idle_cpu, label="idle")
    plt.xlabel("Time (seconds)")
    plt.ylabel("CPU %")
    plt.title("CPU Usage Over Time")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.savefig("cpu_usage.png")
    print("Saved CPU plot to cpu_usage.png")
    plt.close()

def plot_memory(mem_data):
    if not mem_data:
        print("No memory data found.")
        return

    times = [r[0] for r in mem_data]
    total = [r[1] for r in mem_data]
    used = [r[2] for r in mem_data]
    free = [r[3] for r in mem_data]

    plt.figure(figsize=(10, 6))
    plt.plot(times, used, label="used")
    plt.plot(times, free, label="free")
    plt.xlabel("Time (seconds)")
    plt.ylabel("Memory (MB)")
    plt.title("Memory Usage Over Time")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.savefig("memory_usage.png")
    print("Saved memory plot to memory_usage.png")
    plt.close()

if __name__ == "__main__":
    nmon_file = "v6_exp3_drainless_int8_profile_20260601_202251.nmon"

    cpu_data, mem_data = parse_nmon(nmon_file)
    print(f"Parsed CPU samples: {len(cpu_data)}, Memory samples: {len(mem_data)}")

    plot_cpu(cpu_data)
    plot_memory(mem_data)
```
