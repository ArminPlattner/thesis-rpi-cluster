```python
import os
import matplotlib.pyplot as plt

# =====================================================================
# 🛠️ CONFIGURATION
# =====================================================================
# Time interval in seconds between each nmon snapshot
SAMPLING_INTERVAL = 2  

nodes = {
    "Pi 1 (Ingestion)": "pi1_stress_test.nmon",
    "Pi 2 (Kafka)": "pi2_stress_test.nmon",
    "Pi 3 (Inference)": "pi3_stress_test.nmon"
}

colors = {
    "Pi 1 (Ingestion)": "#2ca02c", 
    "Pi 2 (Kafka)": "#1f77b4", 
    "Pi 3 (Inference)": "#d62728"
}

# Create two subplots: top for CPU, bottom for Memory
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(11, 8), dpi=300, sharex=True)

print("Parsing cluster telemetry profiles...")
max_mem = 0

for label, filename in nodes.items():
    if not os.path.exists(filename):
        print(f"⚠️ Warning: {filename} not found.")
        continue
        
    timestamps, cpu_util, mem_used = [], [], []
    
    with open(filename, 'r') as f:
        for line in f:
            if line.startswith("CPU_ALL,T"):
                parts = line.strip().split(',')
                # Calculate real elapsed time: snapshot index * interval
                t_index = int(parts[1][1:])
                elapsed_time = t_index * SAMPLING_INTERVAL
                
                total_cpu = float(parts[2]) + float(parts[3])
                
                timestamps.append(elapsed_time)
                cpu_util.append(total_cpu)
            elif line.startswith("MEM,T"):
                parts = line.strip().split(',')
                val = float(parts[6]) # MEM Used MB
                mem_used.append(val)
                if val > max_mem: max_mem = val

    # Plot CPU
    ax1.plot(timestamps, cpu_util, label=label, color=colors[label], linewidth=2)
    # Plot Memory
    ax2.plot(timestamps[:len(mem_used)], mem_used, label=label, color=colors[label], linewidth=2)

# =====================================================================
# 📈 STYLING
# =====================================================================
# CPU Plot Styling
ax1.set_ylabel("CPU Utilization (%)")
ax1.grid(True, linestyle="--", alpha=0.5)
ax1.legend(loc="upper right", frameon=True)

# Memory Plot Styling
ax2.set_ylabel("Memory Used (MB)")
ax2.set_xlabel("Elapsed Simulation Time (Seconds)")
ax2.set_ylim(0, max_mem * 1.1)
ax2.grid(True, linestyle="--", alpha=0.5)
ax2.legend(loc="lower right", frameon=True)

plt.tight_layout()
plt.savefig("cluster_combined_performance.png", dpi=300)
print(f"📊 Success! Graph saved. Peak memory detected: {max_mem:.2f} MB")
```
